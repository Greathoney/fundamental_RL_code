```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 2
b = 30
next_state_maxQ = np.ones(b)
next_state_prob = np.ones(b) / b
```

```python
Q1 = 0
log1 = []

for n in range(N):
    q = 0
    for i in range(b):
        q += next_state_prob[i] * next_state_maxQ[i]
        log1.append(1-Q1)
    Q1 = q
```

```python
Q2 = 0
log2 = []

for n in range(N):
    for i in range(b):
        Q2 += 0.1 * (next_state_maxQ[i] - Q2)
        log2.append(1-Q2)
```

```python
plt.plot(range(1, b*N+1), log1, label="expected renewal")
plt.plot(range(1, b*N+1), log2, label="sample renewal")
# plt.xticks(range(0, b*N, 2))
plt.xlabel('Number of calculating Q')
plt.ylabel('Error')
plt.legend()
plt.grid()
plt.show()
```
